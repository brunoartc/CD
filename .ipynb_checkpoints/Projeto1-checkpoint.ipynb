## Introdução

O PNAD eh um banco de dados, que reune informacoes do brasil inteiro, e meu projeto visa demonstrar o quao importante é o acesso a internet para os brasileiros, utilizando de varias variaveis quantitativas e qualitativas para verificar desde qual renda que promove o uso de internet ate o que as pessoas deixam de fayer por causa da internet.

## Análise e Resultados

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
def region(DF, ini, fim):
    redxinternet = DF[(DF.RendimentoFamiliar<999999999) & (DF.UF <= fim) & (DF.UF >= ini)]
    return [redxinternet[redxinternet.internetCasa == "sim"].shape[0],redxinternet[redxinternet.internetCasa == "nao"].shape[0]]
```

```python
def allregions(x,DF):
    return [region(DF,11,17)[x], region(DF,21,29)[x], \
    region(DF,31,35)[x], region(DF,41,43)[x], \
    region(DF,50,53)[x]]
```

```python
def graph(DF,name="n sei"):
    objects = ('Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-oeste')
    #y_pos = np.arange(len(objects))
    #plt.xticks(y_pos, objects)
    #print(str(DF))
    #plt.axis([-1,5,0,25000])
    plt.figure(figsize=(20,5))
    plt.subplot(131)
    y_pos = np.arange(len(objects))
    plt.xticks(y_pos, objects)
    plt.axis([-1,5,0,25000])
    plt.title("tem internet em casa (" + name + ")")
    plt.bar(y_pos, allregions(1,DF), align='center', color="green")
    plt.subplot(132)
    y_pos = np.arange(len(objects))
    plt.xticks(y_pos, objects)
    plt.title("nao tem internet em casa (" + name + ")")
    plt.axis([-1,5,0,25000])
    plt.bar(y_pos, allregions(0,DF), align='center', color="red")
```

```python
#Se usar o arquivo descompactado
#pd.read_csv('DOM2015.csv',sep=',')

#Se usar o arquivo compactado como está no dropbox púbico
db2013 = pd.read_csv('DOM2013csv.zip', compression='zip', header=0, sep=',', quotechar='"')
db2014 = pd.read_csv('DOM2014csv.zip', compression='zip', header=0, sep=',', quotechar='"')
db2015 = pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
totalDeMoradores = db2013.V0105
totalDeDomicilios = db2013.V0202
uf = db2013.UF
tvacabo = db2013.V02273.astype('category')
tvacabo.cat.categories = ("sim","nao","nao se sabe")
RendimentoFamiliar= db2013.V4614

agua = db2013.V0211.astype('category')
agua.cat.categories = ("sim","nao")
internetXG = db2013.V02424.astype('category')
internetXG.cat.categories = ("sim","nao")
internetCasa = db2013.V02322.astype('category')
internetCasa.cat.categories = ("sim","nao")


DF2013 = pd.DataFrame()
DF2013["totalDeMoradores"]=totalDeMoradores
DF2013["totalDeDomicilios"]=totalDeDomicilios
DF2013["tvaCabo"]=tvacabo
DF2013["RendimentoFamiliar"]=RendimentoFamiliar
DF2013["agua"]=agua
DF2013["internetXG"]=internetXG
DF2013["internetCasa"]=internetCasa
DF2013["internetCasa"]=internetCasa
DF2013["UF"]=uf

DF2013[DF2013.tvaCabo=="sim"].dropna().head()
```

```python
totalDeMoradores = db2014.V0105
totalDeDomicilios = db2014.V0202
uf = db2014.UF
tvacabo = db2014.V02273.astype('category')
tvacabo.cat.categories = ("sim","nao","nao se sabe")
RendimentoFamiliar= db2014.V4614

agua = db2014.V0211.astype('category')
agua.cat.categories = ("sim","nao")
internetXG = db2014.V02424.astype('category')
internetXG.cat.categories = ("sim","nao")
internetCasa = db2014.V02322.astype('category')
internetCasa.cat.categories = ("sim","nao")


DF2014 = pd.DataFrame()
DF2014["totalDeMoradores"]=totalDeMoradores
DF2014["totalDeDomicilios"]=totalDeDomicilios
DF2014["tvaCabo"]=tvacabo
DF2014["RendimentoFamiliar"]=RendimentoFamiliar
DF2014["agua"]=agua
DF2014["internetXG"]=internetXG
DF2014["internetCasa"]=internetCasa
DF2014["UF"]=uf

DF2014[DF2014.tvaCabo=="sim"].dropna().head()
```

```python
totalDeMoradores = db2015.V0105
totalDeDomicilios = db2015.V0202
uf = db2015.UF
tvacabo = db2015.V02273.astype('category')
tvacabo.cat.categories = ("sim","nao","nao se sabe")
RendimentoFamiliar= db2015.V4614

agua = db2015.V0211.astype('category')
agua.cat.categories = ("sim","nao")
internetXG = db2015.V02424.astype('category')
internetXG.cat.categories = ("sim","nao")
internetCasa = db2015.V02322.astype('category')
internetCasa.cat.categories = ("sim","nao")


DF2015 = pd.DataFrame()
DF2015["totalDeMoradores"]=totalDeMoradores
DF2015["totalDeDomicilios"]=totalDeDomicilios
DF2015["tvaCabo"]=tvacabo
DF2015["RendimentoFamiliar"]=RendimentoFamiliar
DF2015["agua"]=agua
DF2015["internetXG"]=internetXG
DF2015["internetCasa"]=internetCasa
DF2015["UF"]=uf

DF2015[DF2015.tvaCabo=="sim"].dropna().head()
```

## Um estudo sobre a disponibilidade de internet em cada regiao

## num aspecto geral como a renda influencia no acesso a internet

```python
redxinternet = DF2013[DF2013.RendimentoFamiliar<999999999]
slices=(redxinternet["RendimentoFamiliar"].max()-redxinternet["RendimentoFamiliar"].min())/(len(redxinternet["RendimentoFamiliar"])**(1/2))
faixa=np.arange(0,15000,slices)
redxinternet = pd.cut(redxinternet["RendimentoFamiliar"], bins=faixa, right=False)
teste = pd.crosstab(redxinternet,DF2013["internetCasa"])
teste = teste[(teste.sim>0) | ( teste.nao>0 )]
faixa=np.arange(0,2,slices)
plot = teste.plot(kind="bar", title="Renda x acesso a internet 2013")

redxinternet = DF2014[DF2014.RendimentoFamiliar<999999999]
#slices=(redxinternet["RendimentoFamiliar"].max()-redxinternet["RendimentoFamiliar"].min())/(len(redxinternet["RendimentoFamiliar"])**(1/2))
faixa=np.arange(0,15000,slices)
redxinternet = pd.cut(redxinternet["RendimentoFamiliar"], bins=faixa, right=False)
teste = pd.crosstab(redxinternet,DF2014["internetCasa"])
teste = teste[(teste.sim>0) | ( teste.nao>0 )]
faixa=np.arange(0,2000,slices)
plot = teste.plot(kind="bar", title="Renda x acesso a internet 2014")

redxinternet = DF2015[DF2015.RendimentoFamiliar<999999999]
#slices=(redxinternet["RendimentoFamiliar"].max()-redxinternet["RendimentoFamiliar"].min())/(len(redxinternet["RendimentoFamiliar"])**(1/2))
faixa=np.arange(0,15000,slices)
redxinternet = pd.cut(redxinternet["RendimentoFamiliar"], bins=faixa, right=False)
teste = pd.crosstab(redxinternet,DF2015["internetCasa"])
teste = teste[(teste.sim>0) | ( teste.nao>0 )]
faixa=np.arange(0,2000,slices)
plot = teste.plot(kind="bar", title="Renda x acesso a internet 2015")

```

## O QUE AS PESSOAS ACHAM DA TV A CABO QUANDO SE TEM INTERNET AO LONGO DOS ANOS


```python
# Acesso a internet e TV 2013
cross = pd.crosstab(DF2013.tvaCabo,DF2013.internetCasa)
#print(cross)
plt.figure(figsize=(20,5))
plt.subplot(131)
plot = cross["sim"].plot(kind="pie", title="Com internet 2013")
plt.subplot(132)
plot = cross["nao"].plot(kind="pie",  title="Sem internet 2013")



# Acesso a internet e TV 2014
cross = pd.crosstab(DF2014.tvaCabo,DF2014.internetCasa)
#print(cross)
plt.figure(figsize=(20,5))
plt.subplot(131)
plot = cross["sim"].plot(kind="pie", title="Com internet 2014")
plt.subplot(132)
plot = cross["nao"].plot(kind="pie",  title="Sem internet 2014")


# Acesso a internet e TV 2015
cross = pd.crosstab(DF2015.tvaCabo,DF2015.internetCasa)
#print(cross)
plt.figure(figsize=(20,5))
plt.subplot(131)
plot = cross["sim"].plot(kind="pie", title="Com internet 2015")
plt.subplot(132)
plot = cross["nao"].plot(kind="pie",  title="Sem internet 2015")

```

```python
# Acesso a internet 3ge TV 2013
cross = pd.crosstab(DF2013.tvaCabo,DF2013.internetXG)
#print(cross)
plt.figure(figsize=(20,5))
plt.subplot(131)
plot = cross["sim"].plot(kind="pie", title="Com internet 3g 2013")
plt.subplot(132)
plot = cross["nao"].plot(kind="pie",  title="Sem internet 3g 2013")



# Acesso a internet 3ge TV 2014
cross = pd.crosstab(DF2014.tvaCabo,DF2014.internetXG)
#print(cross)
plt.figure(figsize=(20,5))
plt.subplot(131)
plot = cross["sim"].plot(kind="pie", title="Com internet 3g 2014")
plt.subplot(132)
plot = cross["nao"].plot(kind="pie",  title="Sem internet 3g 2014")


# Acesso a internet 3ge TV 2015
cross = pd.crosstab(DF2015.tvaCabo,DF2015.internetXG)
#print(cross)
plt.figure(figsize=(20,5))
plt.subplot(131)
plot = cross["sim"].plot(kind="pie", title="Com internet 3g 2015")
plt.subplot(132)
plot = cross["nao"].plot(kind="pie",  title="Sem internet 3g 2015")

```
